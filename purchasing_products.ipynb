```python
import json
import time
import uuid


#Opening the records.json file
#And, after reading data from this file
#Storing this data into the dictionary inventory
fd = open("records.json","r")

records_data = fd.read()

fd.close()

inventory = json.loads(records_data)




#Opening the sales.json file
#And, after reading data from this file
#Storing this data into the dictionary sales
fd = open("sales.json","r")

sales_data = fd.read()

fd.close()

sales = json.loads(sales_data)




#Printing the inventory
print("\nProducts available in the inventory are:\n")

print("Id\t\tName\t\t\tQuantity\tExpiry date\tMRP\tDiscount")

print("------------------------------------------------------------------------------------------\n")

for prod in inventory:
    print(prod.ljust(16),end="")
    print(inventory[prod]['Name'].ljust(27),end="")
    print(str(inventory[prod]['Quantity']).ljust(14),end="")
    print(inventory[prod]['Expiry Date'].ljust(15),end="")
    print(str(inventory[prod]['MRP']).ljust(10),end="")
    print(str(inventory[prod]['Discount']).ljust(7),end="")
    print("\n")
    


#Asking the user to enter the details of product he/she want to buy    
name = input("Enter your name: ")
prod_id = input("Enter the product id of the product you want to buy: ")
prod_quantity = int(input("Enter the quantity of the product you want to buy: "))


#Checking if product user wants to buy is available in inventory or not
if(prod_id in inventory.keys()):
    
    #Checking if enough quantity of the product is available or not
    if(prod_quantity <= inventory[prod_id]['Quantity']):
        #Updating the Quantity of the Product
        inventory[prod_id]['Quantity'] = inventory[prod_id]['Quantity'] - prod_quantity
        mrp = inventory[prod_id]['MRP']
        discount = (inventory[prod_id]['Discount']/100)
        discounted_mrp = mrp - mrp*discount
        price = mrp * prod_quantity
        discounted_price = discounted_mrp*prod_quantity
        
        #Inserting the transaction into the sales dictionary
        sales[str(uuid.uuid1())] = {'Product_Id': prod_id, 'Costumer Name': name, 'Quantity': prod_quantity, 'Price': discounted_mrp, 'Grand Total': discounted_price, 'Time': time.ctime()}
        
        #Printing the Bill
        print("\n------------------------------------------------------*************---------------------------------------------------------\n")
        print("Your bill is:\n")
        print(time.ctime())
        print("\nHello ",name,":\n")
        print("Id\t\tName\t\t\tQuantity\tMRP\tDiscount\tPrice\t\tDiscounted Price")
        print("-----------------------------------------------------------------------------------------------------------------------------\n")


        print(prod_id.ljust(16),end="")
        print(inventory[prod_id]['Name'].ljust(27),end="")
        print(str(prod_quantity).ljust(13),end="")
        print(str(inventory[prod_id]['MRP']).ljust(10),end="")
        print(str(inventory[prod_id]['Discount']),"%".ljust(12),end="")
        print(str(price).ljust(17),end="")
        print(str(discounted_price).ljust(7),end="")
        print("\n")

        print("\n\n\t\t\t\t\t\t\t\t\t\t\t        Grand Total: ",discounted_price)
        print("\n\n\nThanks for shopping with us............")
        print("\n------------------------------------------------------*************---------------------------------------------------------\n")
    
    #If enough quantity of product is not available
    else:
        print("Enough quantity is not available.")
        
#If product is not available in the inventory
else:
    print("Product is not available in the inventory.")
        
        
        

        

#Opening the records.json and sales.json files in write mode
#And, updating them
js = json.dumps(inventory,indent=2)
fd = open("records.json",'w')
fd.write(js)
fd.close()


js = json.dumps(sales,indent=2)
fd = open("sales.json",'w')
fd.write(js)
fd.close()
```
